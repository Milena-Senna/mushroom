# porto il carattere '?' a nan

```python
fungo.replace('?', np.nan , inplace=True)
```

```python
fungo[['stalk-root']]
```

```python
fungo.isnull().sum()
```

```python
#imputazione colonna con valori categoriali e nan, valore più frequente assegnato ai valori nulli
fungo_più_frequente = fungo[['stalk-root']].apply(lambda x: x.fillna(x.value_counts().index[0]))
fungo_più_frequente
```

```python
#controllo che il valore più frequente è la 'b'
fungo['stalk-root'].value_counts()
```

```python
#colonna 'stalk-root' di nuovo interamente object e senza valori nulli
fungo['stalk-root'] = fungo_più_frequente.astype(str).copy()
```

```python
fungo['stalk-root']
```

# Da caratteri a valori

```python
s = (fungo.dtypes == 'object')
fungo_object_cols = list(s[s].index)
```

```python
#creo copia per evitare di modificare dati originali
label_fungo = fungo.copy()

# Applica il codificatore di etichette a ogni colonna con dati categorici
label_encoder = LabelEncoder()
for col in fungo_object_cols:
    label_fungo[col] = label_encoder.fit_transform(fungo[col])
```

```python
label_fungo
```

```python
fungo
```

# Splitto

```python
y = label_fungo.class_edible
X = label_fungo.drop('class_edible',axis=1,inplace=False)
```

```python
#il target
y
```

```python
X
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

# Modellazione e calcolo accuracy

```python
# definizione del modello
fungo_model = DecisionTreeRegressor(random_state=1)
# alleno il modello
fungo_model.fit(train_X, train_y)
```

```python
val_predictions = fungo_model.predict(val_X)
```

```python
#accuracy
accuracy_fungo = accuracy_score(val_y, val_predictions)*100
accuracy_fungo
```
